```python
genre_collection = list(genre_df['Genres'])
print(genre_collection[0:20])
genre_set = set()
for gg in genre_collection:
    try:
        gglist = gg.split("; ")
    except AttributeError:
        continue
    genre_set.update(gglist)
genres = list(genre_set)
genres
```

```python
def create_vector(row):
    #print(row)
    glabels = row['Genres']
    try:
        glabels = glabels.split("; ")
    except AttributeError:
        glabels = []
    indices = [genres.index(x) for x in glabels]
    array = [0] * len(genres)
    for i in indices:
        array[i] = 1
    return pd.Series(array, index=genres)
```

```python
vector_genres = genre_df.apply(create_vector, axis=1)
```
